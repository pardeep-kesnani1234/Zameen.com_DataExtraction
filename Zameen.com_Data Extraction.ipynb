```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#url = 'https://www.zameen.com/index/buy/residential_property/residential_property-pakistan-price-index.html'
#url = 'https://www.zameen.com/index/buy/plots/plots-pakistan-price-index.html'
url = 'https://www.zameen.com/index/buy/houses/houses-pakistan-price-index.html'
response = requests.get(url)
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup
```

```python
soup.findAll("section",class_='residential history-detail mb20')[0].findAll('ul')[2:]
```

```python
list1 = []
list2 = []
for i in soup.findAll("section",class_='residential history-detail mb20')[0].findAll('ul')[2:]:
    print(str(i.text.strip().replace('\n',','))+"break")
    #print(list(i.text.strip().replace('\n',','))
    list1 = i.text.strip()
    list2.append(list1)
    #list2.append()
    
```

```python
list2[0].split('\n')[2]
```

```python
list3 = []
for i in list2:
    print(i.split('\n'))
    list3.append(i.split('\n'))
    print(list3)
```

```python
month_year = []
index_range = []
price_sqft = []
for j in list3:
    print(j)
    month_year.append(j[0])
    index_range.append(j[1])
    price_sqft.append(j[2])
```

```python
final_df =  pd.DataFrame([month_year,index_range,price_sqft],index=['Month','Index','Price/Sq. Ft.']).transpose()
```

```python
final_df
```

```python
final_df.to_csv('house_pakistan_price.csv')
```
